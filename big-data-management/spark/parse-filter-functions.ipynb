```
# Import necessary libraries
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, avg, broadcast
from pyspark.sql.window import Window
import pyspark.sql.functions as F

# Create a Spark session
spark = SparkSession.builder.appName("Homework4").getOrCreate()

# Load input data
temperature_filepath = "dbfs:/FileStore/shared_uploads/ptw190000@utdallas.edu/city_temperature.csv"
country_list_filepath = "dbfs:/FileStore/shared_uploads/ptw190000@utdallas.edu/country_list.csv"

# create DF
city_temp_df = spark.read.csv(temperature_filepath, header=True)
# Remove -99 temps
city_temp_df = city_temp_df.filter(city_temp_df["AvgTemperature"] != -99)

country_list_df = spark.read.option("header", "true").option("quote", "\"").csv(country_list_filepath)
```

## Part A

```
# A: Find the average of AvgTemperature for each Region
# Group by region, aggregate the avg of avgtemps, and order by region alphabetically
avg_temp_by_region = city_temp_df.groupBy("Region").agg(avg("AvgTemperature").alias("AvgTemperature")).orderBy("Region")
avg_temp_by_region.display()
```

## Part B

```
# B: Find the average of AvgTemperature by Year for countries only located in the “Asia” Region
# Filter by Asia, and select only necessary columns
asia_temp_df = city_temp_df.filter(col("Region") == "Asia").select("Country", "Year", "AvgTemperature")

# group by country and year, and aggregate the temps
avg_temp_asia = asia_temp_df.groupBy("Country", "Year").agg(avg("AvgTemperature").alias("AvgTemp"))

# Sort the temps
sorted_avg_temp_asia = avg_temp_asia.orderBy("Country", "Year")
sorted_avg_temp_asia.display()
```

## Part C

```
# C: Find the average of AvgTemperature by City only located in the Country “Spain”
# Filter where country is spain, group by city, and aggregate avg temps.
avg_temp_spain_cities = city_temp_df.filter(col("Country") == "Spain").groupBy("City").agg(avg("AvgTemperature").alias("AvgTemperature"))
avg_temp_spain_cities.display()
```

## Part D

```
# D: For each country, find the capital and average of AvgTemperature of that capital city
# First, join city_temp_df with country_list_df to get the capital for each country
joined_df = country_list_df.alias("cl").join(city_temp_df.alias("c"), col("c.City") == col("cl.capital"), "inner")
# Then, calculate the average temperature for each capital
avg_temp_capital = joined_df.groupBy("cl.Country", "cl.capital").agg(avg("c.AvgTemperature").alias("AvgTemperature"))
avg_temp_capital = avg_temp_capital.sort("Country")
avg_temp_capital.display()
```

## Part E

```
# Task 1E: Solve Task 1D using a broadcast variable
# Broadcast the smaller DataFrame (country_list_df) for efficient join
broadcast_country_list_df = broadcast(country_list_df)
# Use the broadcast variable for joining
joined_broadcast_df = city_temp_df.alias("c").join(broadcast_country_list_df.alias("bcl"), col("c.City") == col("bcl.capital"), "inner")
avg_temp_capital_broadcast = joined_broadcast_df.groupBy("bcl.Country", "bcl.capital").agg(avg("AvgTemperature").alias("AvgTemperature"))
avg_temp_capital_broadcast = avg_temp_capital_broadcast.sort("Country")

avg_temp_capital_broadcast.display()
```

## Part F

```
# Task 1F: Solve Task 1D using User Defined Functions (UDF)
from pyspark.sql.functions import udf, col, avg
from pyspark.sql.types import StringType, IntegerType, BooleanType, DoubleType

# Define UDF to filter rows for Year >= 2000
def filter_data(year):
    return int(year) >= 2000  # Cast year to integer

# Define UDF to display output
def display_output(capital, country, avg_temp):
    return f"{capital} is the capital of {country} and its average temperature is {avg_temp}"

# Register UDFs
filter_years_udf = udf(filter_data, BooleanType())
display_output_udf = udf(display_output, StringType())

# Filter the data for Year >= 2000 using the filter_years_udf
filtered_data = joined_df.filter(filter_years_udf(col("Year"))) 

# Calculate the average temperature for each capital
avg_temp_capital = filtered_data.groupBy("c.Country", "Capital").agg(avg(col("AvgTemperature")).alias("AvgTemperature"))

# Apply the display_output_udf to create the final result column
final_result = avg_temp_capital.withColumn("Result", display_output_udf(col("Capital"), col("Country"), col("AvgTemperature"))) 

final_result.display()
```
